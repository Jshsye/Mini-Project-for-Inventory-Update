Mini Project for Inventory Update:

```python
class Product():
    def __init__(self, id, name, price, stock):
        self.id= id
        self.name= name
        self.price= price
        self.stock = stock

    def display(self):
        print(f"Id: {self.id}, name: {self.name}, price: {self.price}, stock: {self.stock}")

class Inventory:
    def __init__(self):
        self.products= []

    def add_product(self, product):
        self.products.append(product)
        print(f'Product "{product.name}" add successfully.')

    def remove_product(self, product_id):
        for p in self.products:
            if p.id== product_id:
                self.products.remove(p)
                print(f'Product {p} Remove Sucessfully.')
                return
            print('Product Not Found!')

    def search_product(self, product_id):
        for p in self.products:
            if p.id== product_id:
                print(f"Product Found!.")
                p.display()
                return p
            print('Product not Found!')
            return None
        
    def update_stock(self, product_id, quantity_sold):
        for p in self.products:
            if p.id == product_id:
                if p.stock >= quantity_sold:
                    p.stock -= quantity_sold
                    print(f'Stock Updated. Remaining stock {p.stock}')
                else:
                    print('Not Enough Stock Available')

                return
            print('Product No Found!')


    def display_all(self):
        print('\n All Products in Inventory: ')
        for p in self.products:
            p.display()


# Adding Data
store= Inventory()

# Another Storage [warehouse, gudan, inventory]

warehouse= Inventory()
p1= Product(1, 'Laptop', 50000, 10)  
p2= Product(2, 'Mobile', 15000, 21)
p3= Product(3, 'Mouse', 250, 300)

# Store data:
store.add_product(p1)
store.add_product(p2)
store.add_product(p3)
warehouse.add_product(p1)


store.remove_product(1) # Remove Data with Product ID

store.add_product(p1) # Storing Added Data

store.search_product(1) # Searching Product Using Product_id

store.display_all() # Shows all the stock data available

warehouse.display_all()

warehouse.search_product(2)
```
